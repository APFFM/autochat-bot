```python
def check_doc_uniqueness(list_of_dicts):
    titles_set = set()
    is_unique = True

    for dictionary in list_of_dicts:
        title = dictionary.get('title')
        if title in titles_set:
            is_unique = False
            break
        else:
            titles_set.add(title)

    if is_unique:
        print("All 'title' values are unique.")
    else:
        print("Duplicate 'title' values found.")
        
def check_doc_uniqueness_full(full_dict):
    titles_set = set()
    is_unique = True
    title_list = []
    recurring_title_list = []
    
    print("Checking full uniqueness")

    for key, val in full_dict.items():
        print(f"Now at ({key})")
        #print(type(val))
        #print(val[0])
        for dictionary in val:
            title = dictionary.get('title')
            title_list.append(title)
            if title in titles_set:
                recurring_title_list.append(title)
                is_unique = False
            else:
                titles_set.add(title)
        print(f"Current title_set length: {len(titles_set)}")

    if is_unique:
        print("All 'title' values are unique.")
    else:
        print("Duplicate 'title' values found.")
        
    return title_list, recurring_title_list
```

```python
for key in first_dict.keys():
    print(f"Uniquness test for category ({key}):")
    check_doc_uniqueness(first_dict[key])
    
for key in second_dict.keys():
    print(f"Uniquness test for category ({key}):")
    check_doc_uniqueness(second_dict[key])
```

```python
def is_alphanumeric(text):
    # Remove any whitespace from the text
    cleaned_text = text.replace(" ", "")
    
    # Check if the cleaned text is alphanumeric
    return cleaned_text.isalnum()
```

```python
def preprocess_text(text):
    # Define a regular expression pattern to match non-alphanumeric characters
    # The pattern [^\w\s] matches any character that is not a word character (alphanumeric) or whitespace
    pattern = r'[^a-zA-Z0-9\s]'
    
    # Use the re.sub() function to replace matched characters with an empty string
    cleaned_text = re.sub(pattern, ' ', text)
    cleaned_text = re.sub('_', ' ', cleaned_text)
    cleaned_text = re.sub(r'\s+', ' ', cleaned_text)
    
    return cleaned_text
```

```python
with open("./script_jsons/all_docs.json", "r") as f:
    topic_documents = json.load(f)
```

```python
# Iterate through the list of dictionaries and check each summary
alphanumeric_doc_count = 0
count = 0
non_alphanumeric_docs = []
for key in topic_documents.keys():
    for document_info in topic_documents[key]:
        #print(document_info)
        document_info["summary"] = preprocess_text(document_info["summary"])
        #print(summary)
        
        if is_alphanumeric(document_info["summary"]):
            alphanumeric_doc_count += 1
            document_info["summary"] = document_info["summary"].strip()
        else:
            count += 1
            non_alphanumeric_docs.append((key, document_info["summary"]))
            if count % 10 == 0:
                print("******* BREAKING PREMPTIVELY ********")

# Print the count of alphanumeric summaries
print(f"Number, Percentage of alphanumeric summaries: {alphanumeric_doc_count, (alphanumeric_doc_count/50000)*100}")
```
